```python
from pprint import pprint
def get_comments(video_ids):
    comments = []
    try:
        response = youtube.commentThreads().list(
            part='snippet',
            videoId=video_ids,
            textFormat='plainText'
        ).execute()
    except Exception:
        print(f"Fallo conseguir comentarios de {video_ids}")
        return []

    for item in response['items']:
        try:
            comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
            author_channel_id = item['snippet']['topLevelComment']['snippet']['authorChannelId']["value"]
            #author_channel_id = comment['authorChannelId']['value']
            comments.append({"video_id": video_ids, "author": author_channel_id, "comment": comment})
        except KeyError as e:
            print(f"Falta algun valor de un comentario, {e}")
            comments.append(None)
            continue
    return comments
```

```python
from tqdm.notebook import tqdm 
comment_video = [] 
for video in tqdm(video_ids): 
    comments = get_comments(video) 
    comment_video.extend(comments)
```
